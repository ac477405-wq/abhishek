```python
import numpy as np
```

```python
import pandas as pd
```

```python
data=pd.read_csv("healthcare_dataset.csv")
```

```python
#1.Find the most common disease.
data.value_counts("Medical Condition").idxmax()
```

```python
#2. Calculate average treatment cost per disease.
data.groupby("Medical Condition")["Billing Amount"].mean()
```

```python
#3.Find total patients per hospital branch.
data.value_counts("Hospital")
```

```python
#4.Identify patients older than 60.
data.loc[data["Age"]>60]["Name"]
```

```python
#5. Find the branch with highest treatment cost.
a=data["Billing Amount"].max()
d=data.loc[data["Billing Amount"]==a]
d["Hospital"]
```

```python
#6. Calculate average age of patients per disease.
data.groupby("Medical Condition")["Age"].mean()
```

```python
#7. Find top 5 highest treatment cost patients.
data.sort_values("Billing Amount",ascending=False).head(5)
```

```python
#8. Count patients by gender.
data.groupby("Gender").size()
```

```python
#9.Identify disease with highest treatment cost.
a=data["Billing Amount"].max()
d=data.loc[data["Billing Amount"]==a]
d["Medical Condition"]
```

```python
#10. Calculate monthly patient admissions.
data["Date of Admission"]=pd.to_datetime(data["Date of Admission"])
data.groupby(data["Date of Admission"].dt.to_period("M")).size()

```
